```python
#Define x and y tuples
Array_x = np.array([0, 1, 2, 7, 8, 9, 15], dtype ='int')
Array_y = np.array([0,  3,  6,  9, 12, 15, 18])

print(f'X = {Array_x}')
print(f'Y = {Array_y}')

#Plot
plt.plot(Array_x, Array_y, 'bo')
```

```python
#plot with grid
plt.grid()

#Plot
plt.plot(Array_x, Array_y, 'bo')
```

```python
#title, xlabel, ylabel, xlim, ylim
plt.grid()
plt.title("Water Flow per Week")
plt.xlabel("Number of Day")
plt.ylabel("$m^3$")
plt.xlim(-1,8)
plt.ylim(0,30)

#Plot
plt.plot(Array_x, Array_y,'bo', label='Drink Water')
plt.legend()
```

```python
#Multiple curves in one plot
x1 = np.array([1, 1.5, 2, 2.5, 3, 3.5, 3.6])
y1 = np.array([7.5, 8, 8.5, 9, 9.5, 10, 10.5])

x2=np.array([8, 8.5, 9, 9.5, 10, 10.5, 11])
y2=np.array([3, 3.5, 3.7, 4, 4.5, 5, 5.2])

plt.plot(x1,y1,'ro-', label = 'Curve 1')
plt.plot(x2,y2,'b^:', label = 'Curve 2')
plt.legend()
```

```python
#Subplots, suptitle, set_title, set_xlabel, set_ylabel
fig, ax = plt.subplots(1,2)
x1 = np.array([1, 1.5, 2, 2.5, 3, 3.5, 3.6])
y1 = np.array([7.5, 8, 8.5, 9, 9.5, 10, 10.5])

x2=np.array([8, 8.5, 9, 9.5, 10, 10.5, 11])
y2=np.array([3, 3.5, 3.7, 4, 4.5, 5, 5.2])

fig.suptitle("Two plots in a column")
ax[0].plot(x1,y1,'ro-', axes=ax[0])
ax[0].set_title("Plot 0")
ax[0].set_xlabel('Km')

ax[1].plot(x2,y2,'b^:')
ax[1].set_title("Plot 1")
ax[1].set_ylabel('Speed')

plt.tight_layout() #To avoid overlapings
```
